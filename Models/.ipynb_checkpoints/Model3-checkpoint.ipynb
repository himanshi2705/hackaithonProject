```python
#converting timestamp into right format 
df2=df1
df2['timestamp']= df2['timestamp'].apply(lambda x: pd.Timestamp(x).minute)
df2.head(50)
```

```python
arr= []

for i in range(0, len(df3['timestamp'].to_numpy())):
    temp=i*5
    arr.append(temp)
    

series= pd.Series(arr)

series
```

```python
df3['timestamp']= series
```

```python
#remove null values from dataset 
df4= df3.dropna()
```

```python
#Irrigation Field has 0 means the field has not been irrigated 
df5= df4.drop(['Irrigation field 1'], axis=1)

df5.size
```

```python
df5.rename(columns= {'Soil humidity 1': 'Soil_humidity', 'Air temperature (C)':'Temperature(C)','Air humidity (%)':'Air_humidity(%)'}, inplace=True)

df5.head()
```

```python
#dependent and independent variable
y= df5['Soil_humidity']
X= df5.drop(['Soil_humidity'], axis='columns')
```

```python
y
```

```python
X
```

# Model Training

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.5,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score(LinearRegression(), X, y, cv=cv)
```

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'normalize': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(X,y)
```

```python

# create a regressor object
regressor = DecisionTreeRegressor(random_state = 0) 
  
# fit the regressor with X and Y data
regressor.fit(X_train, y_train)
regressor.score(X_test,y_test)
```

```python
#we conclude that the accuracy of decision tree regressor is the most and therefore we select Decision Tree

#predic moisture at particular condition and check if it is less than threshold 
prediction = regressor.predict([[192, 26, 50, 101.1, 2.17]] ) #192 minutes means 8 hours 


prediction
```

```python
#mean square error is also less 
from sklearn.metrics import mean_squared_error

pred= regressor.predict(X_test)
print(mean_squared_error(y_test, pred))
```

```python
#suppose threshold is 69
t=69

if(prediction<t):
    print("You should irrigate your field after 8 hours " )
```
